### **> Cell 1: Load Building_2 Data**

Explanation:

In this cell, the building_2 CSV file is loaded. Electricity consumption and solar generation data are stored in separate lists

```python
file = open("building_2.csv", "r")
lines = file.readlines()
file.close()

non_shiftable_load = []
solar_generation_per_kw = []

for line in lines[1:]:
    data = line.strip().split(",")

    non_shiftable_load.append(float(data[1]))      
    solar_generation_per_kw.append(float(data[2])) 
```

### **> Cell 2: Total Electricity Consumption**

Explanation:

This cell calculates the total electricity consumption of the building by taking the sum of all hourly electricity values.

```python
total_consumption = sum(non_shiftable_load)

print("Total Electricity Consumption (kWh):", total_consumption)
```

### **> Cell 3: Convert Solar Generation to Actual Energy**

Explanation:

Solar generation values are provided in watts per kilowatt. In this cell, the values are converted into kilowatt units and stored as net solar generation of Building 2.


```python
net_solar_generation_building_2 = []

for value in solar_generation_per_kw:
    net_solar_generation_building_2.append(value / 1000)
```

### **> Cell 4: Load Electricity Price File**

Explanation:

This cell loads the electricity price data using basic Python file handling and stores the price values in a list.

```python
price_file = open("pricing.csv", "r")
price_lines = price_file.readlines()
price_file.close()

price = []

for line in price_lines[1:]:
    values = line.strip().split(",")        # i am use split for seperation
    for v in values:
        price.append(float(v))

```

### **> Cell 5: Calculate Bill Without PV**

Explanation:

This cell calculates the electricity bill without solar power by multiplying hourly electricity consumption with hourly electricity prices.

```python
bill_without_PV = []

for i in range(len(non_shiftable_load)):
    bill_without_PV.append(non_shiftable_load[i] * price[i])

total_bill_without_PV = sum(bill_without_PV)

print("Total Bill Without PV:", total_bill_without_PV)
```

### **> Cell 6: Net Electricity Consumption**

Explanation:

In this cell, solar generation is subtracted from electricity consumption to find the net electricity consumption for each hour.

```python
net_consumption = []

for i in range(len(non_shiftable_load)):
    net_consumption.append(non_shiftable_load[i] - net_solar_generation_building_2[i])
```

### **> Cell 7: Separate Consumption and Wasted Energy**
Explanation:

Positive values represent electricity taken from the grid, while negative values represent surplus or wasted energy. These are separated into two lists.

```python
net_electricity_consumption_with_PV = []
wasted_energy = []

for value in net_consumption:
    if value >= 0:
        net_electricity_consumption_with_PV.append(value)
    else:
        wasted_energy.append(value)

print("Net Electricity Consumption With PV:", sum(net_electricity_consumption_with_PV))
print("Total Wasted Energy:", sum(wasted_energy))
```

### **> Cell 8: Basic Visualization Using Text Output**
Explanation:

visualization is done by displaying sample values, maximum, minimum, and average values to help interpret the data.

```python
print("Sample Electricity Consumption (first 10 hours):")
print(non_shiftable_load[:10])

print("\nSample Solar Generation (first 10 hours):")
print(net_solar_generation_building_2[:10])

print("\nMaximum Consumption:", max(non_shiftable_load))
print("Minimum Consumption:", min(non_shiftable_load))
print("Average Consumption:", sum(non_shiftable_load)/len(non_shiftable_load))
```

### **> Cell 9: Interpretation (As Output)**

Explanation:

This cell prints a short interpretation of the results

```python
if sum(wasted_energy) < 0:
    print("\nSolar generation reduces net electricity consumption.")
    print("Some energy is surplus and can be stored or exported to the grid.")
else:
    print("\nSolar generation is not sufficient to meet the building demand.")
```
